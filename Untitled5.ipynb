```python

# Import necessary libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
import pytesseract
from google.colab.patches import cv2_imshow
```

```python
# Ensure Tesseract is installed
!sudo apt-get install tesseract-ocr
!pip install pytesseract
```

```python
# Load the image
image = cv2.imread('/content/plate_image.jpg')
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
# Display the original image
plt.figure(figsize=(10, 10))
plt.subplot(2, 3, 1)
plt.title("Original Image")
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')
```

```python
# 1. Image Enhancement by Smoothing
blurred_image = cv2.GaussianBlur(image, (5, 5), 0)
plt.subplot(2, 3, 2)
plt.title("Smoothed Image")
plt.imshow(cv2.cvtColor(blurred_image, cv2.COLOR_BGR2RGB))
plt.axis('off')
```

```python
# 2. Image Enhancement by Sharpening
kernel = np.array([[0, -1, 0], [-1, 5,-1], [0, -1, 0]])
sharpened_image = cv2.filter2D(image, -1, kernel)
plt.subplot(2, 3, 3)
plt.title("Sharpened Image")
plt.imshow(cv2.cvtColor(sharpened_image, cv2.COLOR_BGR2RGB))
plt.axis('off')

```

```python
# 3. Noise Addition
from skimage.util import random_noise
noisy_image = random_noise(image, mode='s&p', amount=0.05)
noisy_image = np.array(255 * noisy_image, dtype='uint8')
plt.subplot(2, 3, 4)
plt.title("Noisy Image")
plt.imshow(cv2.cvtColor(noisy_image, cv2.COLOR_BGR2RGB))
plt.axis('off')
```

```python
# 4. Image Restoration (Denoising)
from skimage.restoration import denoise_bilateral
restored_image = denoise_bilateral(noisy_image, multichannel=True)
restored_image = np.array(255 * restored_image, dtype='uint8')
plt.subplot(2, 3, 5)
plt.title("Restored Image")
plt.imshow(cv2.cvtColor(restored_image, cv2.COLOR_BGR2RGB))
plt.axis('off')

# 5. Image Segmentation (Detect and Extract Number Plate)
# Convert to grayscale and apply edge detection
edges = cv2.Canny(gray_image, 50, 200)
plt.subplot(2, 3, 6)
plt.title("Edges")
plt.imshow(edges, cmap='gray')
plt.axis('off')
```

```python
# Preprocessing
gray_image = cv2.bilateralFilter(gray_image, 11, 17, 17) # Reduce noise while keeping edges sharp
edges = cv2.Canny(gray_image, 30, 200) # Edge detection
```

```python
# Find contours
contours, _ = cv2.findContours(edges, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
contours = sorted(contours, key=cv2.contourArea, reverse=True)[:10]

# Sort contours by area and take the largest ones
number_plate = None
for cnt in contours:
  approx = cv2.approxPolyDP(cnt, 0.02 * cv2.arcLength(cnt, True), True)
  if len(approx) == 4:
    x, y, w, h = cv2.boundingRect(cnt)
    aspect_ratio = w / float(h)
    if 2 <= aspect_ratio <= 6:
       number_plate = gray_image[y:y+h, x:x+w]
       cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 3)
       break

# OCR processing if the number plate is found
if number_plate is not None:
  # Apply additional preprocessing to the number plate region
  number_plate = cv2.GaussianBlur(number_plate, (5, 5), 0)
  _,number_plate_thresh = cv2.threshold(number_plate, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
  # Use Tesseract to extract text from the number plate
  custom_config = r'--oem 3 --psm 8'
  text = pytesseract.image_to_string(number_plate_thresh, config=custom_config)
  # Display results cv2_imshow(image)
  plt.figure(figsize=(5, 5))
  plt.title("Extracted Number Plate")
  plt.imshow(number_plate_thresh, cmap='gray')
  plt.axis('off')

  print("Extracted Number:", text.strip())
else:
  print("Number plate not detected.")
plt.show()
```
